#Etape 3 - Jointure des fichiers

##3.1 - Jointure des fichiers df_da_filtre et df_pop_filtre

```python
#Jointure des fichiers df_da_filtre et df_pop_filtre
df_merge = pd.merge(df_da_filtre, df_pop_filtre, how='outer', on='Code zone', indicator=True)
df_merge
```

```python
#Vérification des right_only
df_merge[df_merge['_merge'] == 'right_only']
```

```python
#Les pays qui des valeurs "right_only" sont des pays où la production de viandes de volailles n'existe pas (ex: Bermudes) ou dont il n'y a pas de données (ex: Andorre).
#Ces pays seront supprimés de df_merge.
```

```python
#Calcul de la porportion de ces pays 'right_only' en terme de population
df_merge[df_merge['_merge'] == 'right_only']['Population'].sum() / df_merge['Population'].sum()*100
```

```python
#Suppression des lignes "right_only" de df_merge
df_merge = df_merge[df_merge['_merge'] != 'right_only']
```

```python
#Vérification
df_merge
```

```python
#Suppression des colonnes inutiles
df_merge = df_merge.drop(columns=['_merge', 'Zone_y'])
```

```python
#Vérification
df_merge
```

```python
#Renommage de la colonne "Zone_x" en "Pays"
df_merge = df_merge.rename(columns={'Zone_x': 'Pays'})
df_merge.head()
```

```python
#Vérification des doublons
df_merge[df_merge.duplicated()]
```

```python
#Vérification du nombre de doublons
df_merge.duplicated().sum()
```

```python
#Vérification du nombre de valeurs manquantes
df_merge.isna().sum()
```

```python
#Vérification des lignes ayant au moins une valeur manquante
df_merge[df_merge.isna().any(axis=1)]
```

```python
#Certains pays n'exportent pas ou les données sur les exportations ne sont pas disponibles.
#Il n'y a que la République démocratique populaire lao dont touces ces informations sont absentes : Disponibilité intérieure, Exportations, Importations, Nourriture, Production, Variation de stock.
#Afin de ne pas laisser ces cases vides, je remplace NaN par 0.
```

```python
#Remplacer les NaN par 0
df_merge = df_merge.fillna(0)
```

```python
#Vérification du nombre de lignes ayant des valeurs manquantes
df_merge[df_merge.isna().any(axis=1)]
```

##3.2 - Jointure des fichiers df_merge et df_pib_filtre

```python
#Jointure des fichiers df_merge et df_pib_filtre
df_merge_2 = pd.merge(df_merge, df_pib_filtre, how='outer', left_on='Pays', right_on='Country Name', indicator=True)
df_merge_2
```

```python
#Vérification des "right_only"
df_merge_2[df_merge_2['_merge'] == 'right_only']
```

```python
#Les lignes "right_only" correspondent aux pays dans lesquels il n'existe pas de productions de volailles ou les pays qui n'ont pas de correspondance dans df_merge.
```

```python
#Suppression des lignes "right_only" de df_merge_2
df_merge_2_filtre = df_merge_2[df_merge_2['_merge'] != 'right_only']
```

```python
#Vérification
df_merge_2_filtre
```

```python
#Vérification des lignes "left_only"
df_merge_2[df_merge_2['_merge'] == 'left_only']
```

```python
#Afin de pouvoir garder ces lignes et ces pays, je remplace les valeurs NaN par 0.
```

```python
#Garder la colonne "Pays"
df_merge_2_filtre["Pays"] = df_merge_2["Pays"]
```

```python
#Exclure la colonne 'Pays' pour ne pas la toucher
num_like = [
    col for col in df_merge_2_filtre.columns
    if df_merge_2_filtre[col].dtype == "object" and col != "Pays"
]
```

```python
#Filtrer sur les lignes "left_only"
mask = df_merge_2_filtre['_merge'] == 'left_only'
```

```python
#Remplacer NaN par 0
df_merge_2_filtre.loc[mask, num_like] = (
    df_merge_2_filtre.loc[mask, num_like]
        .apply(pd.to_numeric, errors='coerce')
        .fillna(0)
        .values
)
```

```python
num_cols = [col for col in df_merge_2_filtre.select_dtypes(include=['number']).columns
            if col != "Pays"]
```

```python
#Remplacer les NaN des colonnes "pib_2016", "pib_2017" et "évolution_pib_%"
cols = ["pib_2016", "pib_2017", "évolution_pib_%"]

df_merge_2_filtre[cols] = df_merge_2_filtre[cols].fillna(0)
```

```python
#Vérifier les lignes "left_only"
df_merge_2_filtre[df_merge_2_filtre['_merge'] == 'left_only']
```

```python
#Vérification des doublons
df_merge_2_filtre[df_merge_2_filtre.duplicated()]
```

```python
#Vérification des valeurs manquantes
df_merge_2_filtre.isna().sum()
```

```python
#Suppression de la colonne "_merge"
df_merge_2_filtre = df_merge_2_filtre.drop(columns=['_merge'])
```

```python
#Suppression de la colonne "Country Name"
df_merge_2_filtre = df_merge_2_filtre.drop(columns=['Country Name'])
```
